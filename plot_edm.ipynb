```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('seatgeek_edm_data_clean.csv')
```

```python
df.head(3)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
ax.scatter(x=df[df['venue.timezone'] ==  'America/Los_Angeles']['score'], 
           y=df[df['venue.timezone'] ==  'America/Los_Angeles']['venue.score'],
          c='b', label='LA', s=40)

ax.scatter(x=df[df['venue.timezone'] ==  'America/Chicago']['score'], 
           y=df[df['venue.timezone'] ==  'America/Chicago']['venue.score'],
          c='g', label='Chicago', s=40)

ax.scatter(x=df[df['venue.timezone'] ==  'America/New_York']['score'], 
           y=df[df['venue.timezone'] ==  'America/New_York']['venue.score'],
          c='r', label='New York', s=40)

plt.legend(loc='upper right')
plt.legend(bbox_to_anchor=(1.2, 1))
plt.title("Venue Score by Artist Score")
plt.xlabel('Artist Score')
plt.ylabel('Venue Score')
plt.show();
```
