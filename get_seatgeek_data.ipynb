```python
import pandas as pd
import numpy as np
import requests
from time import sleep
from pandas.io.json import json_normalize
import timeit
from __future__ import division
import time
```

# Here is the code to get all EDM data from SeatGeek

```python
# Call URL for all electronic artists
# per_page = 5000 because it is the max

all_edm_url = "https://api.seatgeek.com/2/performers?format=json&client_id=NDQxMDA3NXwxNDU5MDI4MzEx&genres.name=electronic&per_page=5000"
r = requests.get(all_edm_url)
json_all_edm = r.json()
df_all_edm = json_normalize(json_all_edm)
all_elect_artists_ids = json_normalize(df_all_edm['performers'][0])
all_elect_artists_ids
```

```python
# make loop to get all ticket information for all edm artists

master_concerts = pd.DataFrame([])

for i in range(all_elect_artists_ids.shape[0]):
    try:
        artist_concert_info = "https://api.seatgeek.com/2/events?performers.id={}&client_id=NDQxMDA3NXwxNDU5MDI4MzEx&per_page=5000".format(
            all_elect_artists_ids['id'][i])
#        print artist_concert_info
        r = requests.get(artist_concert_info)
        json_concert_per_artist = r.json()
        x = json_normalize(json_concert_per_artist)
        if x['events'][0] != []:
            if i % 25 == 0:
                print 'success! {}'.format(i)
            single_artist_df = json_normalize(x['events'][0])
            single_artist_df['artist_name'] = all_elect_artists_ids['short_name'][i]
            master_concerts = master_concerts.append(single_artist_df)
        else: 
            if i % 25 == 0:
                print "no data for {}".format(i)
    except: 
        print 'who knows'
```

```python
master_concerts = master_concerts.reset_index(drop=True)
master_concerts
```

```python
master_concerts.to_csv('seatgeek_edm_data.csv')
```
